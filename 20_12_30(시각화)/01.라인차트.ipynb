```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

plt.xlabel('x축라벨') # x축 이름 설정
plt.ylabel('y축라벨')
plt.grid(True) # 격자 출력
plt.title('타이틀')
plt.plot(x,y1,'bo-', x,y2,'rs--', x,y3,'gp-.')
plt.legend(['국어','영어','수학']) # plot을 한후 작성
plt.show()
```

```python
xname = ['홍길동','이순신','임꺽정','김철수','이황']
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]

plt.figure(figsize=(7,7))
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')
plt.grid(True)
# (x, xname) x를 xname으로 매핑 ex) 1은 홍길동, 2는 이순신 ...
plt.xticks(x,xname,rotation=45) # x축 값의 이름 변경옵션

plt.plot(x,y1,'ro--', alpha=0.5)
plt.text(1,18,'18점') # 해당 지점에 텍스트 표시
 
for n1,n2 in zip(x,y1):
    plt.text(n1,n2, f'{n2}점') # 해당 지점에 텍스트 표시

plt.legend(['국어'])
plt.show()
```

```python
x  = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

# 한번에 여러개 만듬
fig, ax = plt.subplots(figsize=(6,4))
ax.set_xlabel('x축')
ax.set_ylabel('y1축', color = 'blue', rotation=0)
ax.plot(x,y1)
ax2 = ax.twinx() # y축을 오른쪽으로
ax2.set_ylabel('y2축', color ='red')
ax2.plot(x,y2,'r--')
plt.show()
```
