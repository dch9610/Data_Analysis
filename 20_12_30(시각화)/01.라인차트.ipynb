```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

plt.xlabel('x') # x  
plt.ylabel('y')
plt.grid(True) #  
plt.title('')
plt.plot(x,y1,'bo-', x,y2,'rs--', x,y3,'gp-.')
plt.legend(['','','']) # plot  
plt.show()
```

```python
xname = ['','','','','']
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]

plt.figure(figsize=(7,7))
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.grid(True)
# (x, xname) x xname  ex) 1 , 2  ...
plt.xticks(x,xname,rotation=45) # x   

plt.plot(x,y1,'ro--', alpha=0.5)
plt.text(1,18,'18') #    
 
for n1,n2 in zip(x,y1):
    plt.text(n1,n2, f'{n2}') #    

plt.legend([''])
plt.show()
```

```python
x  = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

#   
fig, ax = plt.subplots(figsize=(6,4))
ax.set_xlabel('x')
ax.set_ylabel('y1', color = 'blue', rotation=0)
ax.plot(x,y1)
ax2 = ax.twinx() # y 
ax2.set_ylabel('y2', color ='red')
ax2.plot(x,y2,'r--')
plt.show()
```
