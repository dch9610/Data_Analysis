```python
from bs4 import BeautifulSoup
import urllib.request as REQ # http 통신 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

### 중앙일보 전체기사 xml

```python
jurl = 'https://rss.joins.com/joins_news_list.xml'
resonse = REQ.urlopen(jurl)
```

```python
soup = BeautifulSoup(resonse, 'html.parser') # html.parser(![...] 부분 제거하고 string값만 가져오게하기), html5lib
soup
```

```python
# soup.select('selector')
for item in soup.select('item'):
    print("기사제목: ", item.title.string)
    print("기사내용: ", item.description.string)
    print('===============================')
```

### 기상청 데이터 xml

```python
kurl = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'
response = REQ.urlopen(kurl) 
```

```python
ksoup = BeautifulSoup(response, 'html.parser')
ksoup
```

```python
# 도시만 출력하시오
for item in ksoup.select('location'):
    print(item.city.string)
```

```python
서울
========
날짜:
날씨:
최저:
최고:
```

```python
for item in ksoup.select('location'):
    print(item.city.string)
    print('====================')
    for a in item.select('data'):
        print("날짜: ", a.tmef.string)
        print("날씨: ", a.wf.string)
        print("최저: ", a.tmn.string)
        print("최고: ", a.tmx.string)
        print('----------------------')
        
```

```python
# 리스트 형식으로 만들기
kList=[]
for item in ksoup.select('location'):
    for a in item.select('data'):
        kList.append({'도시':item.city.string,
                     '날짜':a.tmef.string,
                     '날씨':a.wf.string,
                     '최저':int(a.tmn.string),
                     '최고':int(a.tmx.string)})
kList
```

```python
# 데이터프레임으로 만들기
kdf = pd.DataFrame(kList)
kdf
```

```python
# 특정 컬럼을 인덱스로 바꿈
kdf.set_index(['도시','날짜'],inplace=True)
kdf
```

```python
# 전체 다 보이게 하는 옵션
pd.set_option('display.max_rows',None)
kdf
```

```python
# 6. 느낌 컬럼을 추가하고 최저기온이 
# -10 도 이하면: 매우 추움
# -5~-9 면: 추움
# 나머지 : 보통
# 표시하시요
```

```python
def fn(v):
    if v <= -10:
        return '매우 추움'
    elif -9 <= v <= -5:
        return '추움'
    else:
        return '보통'
```

```python
kdf['느낌'] = kdf['최저'].apply(fn)
kdf
```

#### 9. 여주지역 전체 최고기온 중 최고기온이 0 도 이상이 차지하는비율을 구하시요.
- (예를들어  출력결과:0도이상 15% )


```python
print(yu['최고'].count())
print(yu[yu['최고'] >= 0]['최고'].count())
yu1 = yu['최고'].count()
yu2 = yu[yu['최고'] >= 0]['최고'].count()

print('최고기온이 0도 이상비율: ',round((yu2/yu1)*100,2),'%')

```

```python
a = kdf.loc['여주'].query('최고>=0')['최고'].count()
b = kdf.loc['여주','최고'].count()
print('0도 이상 : ', (a/b*100).round(2),'%')
```

```python
# len을 통해 갯수 파악가능
zeroCount = len(kdf.loc['여주'].query('최고>=0'))
tot = len(kdf.loc['여주'])
print(zeroCount, tot)
round((zeroCount/tot)*100,2)
```

#### 10. 일교차 컬럼을 추가하고,최고기온과 최저기온의 차이가 가장높은 도시,날짜,최저,최고,일교차를 구하시요
