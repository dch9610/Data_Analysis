```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib

# 한글 깨짐 방지
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {'aa':10,'bb':20,'cc':30,'dd':40,'ee':50}
sr = pd.Series(data, name='국어점수')
sr
```

```python
sr.max()
```

```python
# 가장 큰 값의 인덱스 값
sr.idxmax()
```

```python
# 45이하인데이터의 가장큰값을 구하시요.
sr[sr<=45].max()
```

```python
sr.tail(2)
```

```python
# 값이 가장 큰거 출력
sr.nlargest(2)
```

```python
sr.quantile([0.25,0.75])
```

```python
# 중복된 값 제거한 후 값의 종류
sr.unique()
```

```python
# 값의 갯수
sr.value_counts()
```

```python
def fn(v):
    # print('v =',v)
    # print("==============")

    return v+1 if v >30 else v+2
```

```python
sr.apply(fn)
```

```python
sr.apply(lambda v: v+1 if v>30 else v+2)
```

```python
# 국어점수가 40점 이상이면 '합격' 아니면 '불합격'으로 구하시오
sr.apply(lambda v: '합격' if v>=40 else '불합격')
```

```python
# 5개의 범위로 만들어줌
# (9.96, 18.0] : 9.96 < aa <= 18.0
pd.cut(sr,5)
```

```python
pd.cut(sr,5).value_counts().sort_index()
```

```python
pd.cut(sr,[0,20,40,60], labels=['20점','40점','60점'])
```
