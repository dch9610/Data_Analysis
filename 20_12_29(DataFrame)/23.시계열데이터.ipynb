### 시계열 데이터의 정의
- 1. 인덱스가 datetime형
- 2. 시계열데이터의 indexing, slicing을 사용가능`

```python
data1 = [[1,2],[3,4],[5,8],[7,6],[9,12],[11,10],[13,16],[15,20]]
df = pd.DataFrame( data1)
df.index=['2019-01-01','2019-02-01','2019-03-01','2019-04-01',
          '2019-05-01','2019-06-01','2019-07-01','2019-08-01']
# df.index=[datetime(2019,1,1),datetime(2019,2,1),datetime(2019,3,1),datetime(2019,4,1),
#           datetime(2019,4,1),datetime(2019,5,1),datetime(2019,7,1),datetime(2019,8,1)]
df
```

```python
# 시계열 데이터는 타입이 datetime이여야함
df.index
```

### 문자열을 datetime형으로 형변환

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
# ****년**월**일 변환시 에러 발생
df.index=['2019년01월01','2019년02월01','2019년03월01','2019년04월01',
          '2019년05월01','2019년06월01','2019년07월01','2019년08월01']
```

#### 날짜 포맷
- https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavio

```python
# format을 사용하여 맞춰줌
df.index = pd.to_datetime(df.index, format='%Y년%m월%d')
df.index
```

```python
df.columns=['A','B']
df
```

#### 이동평균

```python
# rolling : 이동평균
# window = 값, 값만큼 묶어서 평균을 구함
df.rolling(window=2).mean()
```

```python
# min_periods = 1 Nan값으로 나오는 부분을 평균해서 나오게함
df.rolling(window=3, min_periods=1).mean()
```

```python
plt.xticks(rotation=45)
plt.plot(df.index, df['B'])
plt.plot(df.index, df.rolling(window=2, min_periods=1).mean()['B'])
plt.show()
```

### 주식데이터 가져오기

```python
# data.get_data_yahoo(종목코드, 시작날짜, 끝날짜)
# 삼성전자
samsungDF = data.get_data_yahoo('005930.KS','2017-01-01')
samsungDF
```

```python
samsungDF.info()
```

```python
samsungDF.index
```

```python
# samsungDF.plot(y='Close')
plt.plot(samsungDF.index, samsungDF['Close'])
plt.plot(samsungDF.index, samsungDF.rolling(window=20).mean()['Close'],'r-')
plt.show()
```

#### 시계열데이터만 적용되는 인덱싱 슬라이싱

```python
samsungDF.index
```

```python
# 2019년 데이터 가져오기
samsungDF['2019']
```

```python
samsungDF['2019-01']
```

```python
samsungDF['2019-01':'2019-07']
```

```python
samsungDF.index.year
```

```python
samsungDF['year'] = samsungDF.index.year
```

```python
# 년도별 평균 값
samsungDF.groupby(['year']).mean()
samsungDF.groupby(samsungDF.index.year).mean()
```

```python
# 년도별, 월별
samsungDF.groupby([samsungDF.index.year,samsungDF.index.month]).mean()
```

```python
# resample_rule : 시계열데이터 전용 집계함수
samsungDF.resample('W-SUN').mean() # 일~월 주간별 집계
```

```python
# Q : 분기별 데이터 집합
samsungDF.resample('Q').mean()
```
