```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime 
from pandas_datareader import data

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

plt.rcParams["font.size"] = 13
```

```python
df = pd.read_csv('merge.csv',index_col=False,thousands=',')
df
```

```python
# 필요없는 열 삭제
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df= df.sort_values('지역')
```

```python
df = df.set_index(['지역','국가'])
```

```python
df
```

### 아메리카

```python
ame = df.loc['아메리카']
ame
# for a,b,c in ame['']
```

```python
Cp = []
for n1, n2 in zip(ame['인구수'],ame['환자발생수']):
    Cp.append(round(n2/n1,2)*100)
Cp
```

### 아시아

```python
asia = df.loc['아시아']
asia
```

```python
Cp1 = []
for n1, n2 in zip(asia['인구수'],asia['환자발생수']):
    Cp1.append(round(n2/n1,2)*100)
Cp1
```

### 중동

```python
jung = df.loc['중동']
jung
```

```python
Cp2 = []
for n1, n2 in zip(jung['인구수'],jung['환자발생수']):
    Cp2.append(round(n2/n1,2)*100)
Cp2
```

### 유럽

```python
eu = df.loc['유럽']
eu
```

```python
Cp3 = []
for n1, n2 in zip(eu['인구수'],eu['환자발생수']):
    Cp3.append(round(n2/n1,2)*100)
Cp3
```

## 그래프 표시

```python
plt.figure(figsize=(22,12))
plt.subplot(2,2,1) # 행,열, 넘버
plt.xticks(rotation=90)
plt.title('아메리카 지역')
plt.ylabel('인구대비 확진율')
plt.plot(ame.index,Cp)

plt.subplot(2,2,2) # 행,열, 넘버
plt.xticks(rotation=90)
plt.title('아시아 지역')
plt.ylabel('인구대비 확진율')
plt.plot(asia.index,Cp1)

plt.subplot(2,2,3) # 행,열, 넘버
plt.xticks(rotation=90)
plt.xlabel('중동 지역')
plt.ylabel('인구대비 확진율')
plt.plot(jung.index,Cp2)

plt.subplot(2,2,4) # 행,열, 넘버
plt.xticks(rotation=90)
plt.xlabel('유럽 지역')
plt.ylabel('인구대비 확진율')
plt.plot(eu.index,Cp3)
```

```python
# 인구 대비 확진률로 보아
# 아메리카 지역에서는 미국
# 아시아 지역에서는 몰디브
# 중동 지역에서는 바레인
# 유럽 지역에서는 안도라가 인구 대비 감염이 심각한 것으로 알수 있습니다. 
```
