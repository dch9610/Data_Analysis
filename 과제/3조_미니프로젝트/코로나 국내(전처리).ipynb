```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import urllib.request as REQ

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
curl = 'http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=13&ncvContSeq=&contSeq=&board_id=&gubun='
response = REQ.urlopen(curl)
soup = BeautifulSoup(response, 'html5lib') 
soup
```

```python
ulTag = soup.select('table',{'class':'num midsize'})
ulTag
```

```python
name=[]


for l in ulTag:
    for trs in l.select('tr'):
        name.append(trs.select_one('th').string)
name
```

```python
arr = np.array(name)
arr
```

```python
arr[2:]
```

```python
aa = []

for l in ulTag:
    for trs in l.select('tr'):
            tds4 = trs.select('td:nth-of-type(4)')
            for a in tds4:
                aa.append(a.string)
aa
```

```python
bb = []

for l in ulTag:
    for trs in l.select('tr'):
            tds5 = trs.select('td:nth-of-type(5)')
            for b in tds5:
                bb.append(b.string)
bb
```

```python
cc = []

for l in ulTag:
    for trs in l.select('tr'):
            tds6 = trs.select('td:nth-of-type(6)')
            for c in tds6:
                cc.append(c.string)
cc
```

```python
dd = []

for l in ulTag:
    for trs in l.select('tr'):
            tds7 = trs.select('td:nth-of-type(7)')
            for d in tds7:
                dd.append(d.string)
dd
```

```python
ee = []

for l in ulTag:
    for trs in l.select('tr'):
            tds8 = trs.select('td:nth-of-type(8)')
            for e in tds8:
                ee.append(e.string)
ee
```

```python
koreaDF = pd.DataFrame()
koreaDF
```

```python
koreaDF['확진환자'] = aa
koreaDF['격리중 '] = bb
koreaDF['격리해제  '] = cc
koreaDF['사망자    '] = dd
koreaDF['발생률 '] = ee
koreaDF
```

```python
koreaDF['시도명'] = arr[2:]
koreaDF
```

```python
koreaDF.drop(0, inplace=True)
```

```python
koreaDF.drop(18, inplace=True)
```

# 통계분석

```python
koreaDF.set_index('시도명', inplace=True)
koreaDF
```

```python
koreaDF.info()
```

```python
koreaDF['확진환자'] = koreaDF['확진환자'].apply(lambda x: x.replace(',', ''))
koreaDF
```

```python
koreaDF.columns = ['확진환자', '격리중', '격리해제', '사망자', '발생률']
```

```python
koreaDF
```

```python
def remove_comma(x):
     return x.replace(',', '')

koreaDF['확진환자'] = koreaDF['확진환자'].apply(remove_comma)
koreaDF['격리중'] = koreaDF['격리중'].apply(remove_comma)
koreaDF['격리해제'] = koreaDF['격리해제'].apply(remove_comma)
```

```python
koreaDF
```

```python
koreaDF['확진환자'] = koreaDF['확진환자'].astype(int)
koreaDF['격리중'] = koreaDF['격리중'].astype(int)
koreaDF['격리해제'] = koreaDF['격리해제'].astype(int)
koreaDF['사망자'] = koreaDF['사망자'].astype(int)
koreaDF['발생률'] = koreaDF['발생률'].astype(float)
koreaDF
```

```python
koreaDF.info()
```

```python
koreaDF.to_csv('korea_corona.csv')
```

## 국내 시도별 발생률 - 바차트

```python
koreaDF['발생률'].plot(kind='bar', 
                      title='국내 시도별 발생률', 
                      ylabel='발생률',
                      legend=True, 
                      grid=True, 
                      color='salmon')
plt.show()
```

## 확진환자 중 격리중 환자 비율

```python
koreaDF['격리 비율'] = koreaDF['격리중'] / koreaDF['확진환자']
koreaDF
```

## 확진환자 중 격리해제 환자 비율

```python
koreaDF['격리해제 비율'] = koreaDF['격리해제'] / koreaDF['확진환자']
koreaDF
```

## 확진환자 중 사망자 환자 비율

```python
koreaDF['사망자 비율'] = koreaDF['사망자'] / koreaDF['확진환자']
koreaDF
```

## 지역별 격리 비율, 격리해제 비율, 사망자 비율 - 라인차트

```python
plt.plot(koreaDF['격리 비율'], 'bo-', 
         koreaDF['격리해제 비율'], 'rs--', 
         koreaDF['사망자 비율'], 'yp-.', 
         alpha=0.5)
plt.xlabel('시도명')
plt.ylabel('비율')
plt.grid(True)
plt.title('지역별 비율')
plt.xticks(rotation=90)
plt.legend(['격리 비율', '격리해제 비율', '사망자 비율'])
plt.show()
```
