```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib.request as REQ
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

# 크롤링

##  총 인구수 크롤링

```python
nurl = 'https://ko.wikipedia.org/wiki/%EC%9D%B8%EA%B5%AC%EC%88%9C_%EB%82%98%EB%9D%BC_%EB%AA%A9%EB%A1%9D'
response= REQ.urlopen( nurl)
soup = BeautifulSoup( response, 'html5lib')
soup
```

```python
i = "#mw-content-text > div.mw-parser-output > table "
ulTag = soup.select( i )
print(ulTag)
```

```python
population = list()

for t in ulTag:
    for i in t.select('tr'):
        for e in i.select("td:nth-of-type(1) > a"):
            print(e.string)
            population.append(e.string)
```

```python
number    = list()

for t in ulTag:
    for i in t.select('tr'):
        for e in i.select("td:nth-of-type(3)"):
            print(e.string)
            number.append(e.string)
```

## 국외 데이터 크롤링

```python
kourl='http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=14&ncvContSeq=&contSeq=&board_id=&gubun='
response = REQ.urlopen(kourl)
soup = BeautifulSoup(response, 'html5lib')
soup
```

```python
s = '#content > div > div.box_toggle > div.inner > div.data_table.mgt16 > table > tbody '
ulTag = soup.select_one(s)
ulTag
```

```python
country = []
patient = []
for tr in ulTag.select('tr'):
    print('국가:' , tr.select_one('.w_bold').string)
    print('환자발생수:' , tr.select_one('td:nth-of-type(2)').string)
    country.append({'국가' : tr.select_one('.w_bold').string,
                    '환자발생수': tr.select_one('td:nth-of-type(2)').string})

```

# 총인구 데이터프레임 후 전처리

```python
pop = pd.DataFrame(population,number )
```

```python
pop['인구수'] = pop['인구수'].str.replace(',','')
```

### 다른 데이터와 국가명 맞추기

```python
pop['국가'][0] = '중국'
```

### 데이터 엑셀 저장

```python
pop.to_csv('국가별 총인구.csv')
```

# 국외 데이터 전처리

```python
countryDF = pd.DataFrame(country)
countryDF.set_index('국가', inplace=True)
countryDF = countryDF.dropna()
countryDF
```

### 데이터 값 전처리

### # 지역구분으로 크롤링안된 데이터

```python

countryDF.loc['중국'] = '87,027명(사망 4,634)'
countryDF.loc['이란'] = '1,206,373명(사망 54,814)'
countryDF.loc['미국'] = '18,972,813명(사망 330,644)'
countryDF.loc['이탈리아'] = '2,056,277명(사망 72,370)'
countryDF.loc['호주'] = '28,337명(사망 909)'
countryDF.loc['나이지리아'] = '84,811명(사망 1,264)'
countryDF.loc['기타지역 영토'] = '313,861명(사망 3,762)'

countryDF
```

### 불필요 내용 제거

```python
countryDF['환자발생수'] = countryDF['환자발생수'].map(lambda x:x.replace('\t','').replace('\n',''))
countryDF
```

```python
a=[]
b=[]
for n in range(0,188):
  a.append(countryDF['환자발생수'][n].split('명')[0])
  b.append(countryDF['환자발생수'][n].split('명')[1])
a
b
```

```python
countryDF['환자발생수'] = a
countryDF['사망자수'] = b
```

```python
countryDF['사망자수'] = countryDF['사망자수'].map(lambda x: x.lstrip('(사망'))
countryDF['사망자수'] = countryDF['사망자수'].map(lambda x: x.rstrip(')'))
countryDF
```

### 지역 칼럼 생성

```python
countryDF['지역'] = '_'

countryDF.loc['중국':'타지키스탄']['지역'] = '아시아'
countryDF.loc['이란':'예멘']['지역'] = '중동'
countryDF.loc['미국':'수리남']['지역'] = '아메리카'
countryDF.loc['이탈리아':'러시아']['지역'] = '유럽'
countryDF.loc['호주':'바누아투']['지역'] = '오세아니아'
countryDF.loc['나이지리아':'레소토']['지역'] = '아프리카'
countryDF.loc['기타지역 영토':]['지역'] = '기타'

countryDF.reset_index(drop=False, inplace=True)
```

### nan값 제거

```python
countryDF['사망자수']= countryDF['사망자수'].fillna(0)
```

### 멀티인덱스 설정

```python
countryDF.set_index(['지역','국가'])
```

### 데이터 저장

```python
countryDF.to_csv('country_COVID.csv')
```

# 국내 데이터

```python
koreaDF= pd.read_csv('korea_corona.csv')
koreaDF.drop(columns='Unnamed: 0', inplace=True)
```

```python
koreaDF.set_index('시도명', inplace=True)
koreaDF
```

## 전처리

### 칼럼명 수정

```python
koreaDF.columns = ['확진환자', '격리중', '격리해제', '사망자', '발생률']
```

### 콤마 제거

```python
def remove_comma(x):
     return x.replace(',', '')

koreaDF['확진환자'] = koreaDF['확진환자'].apply(remove_comma)
koreaDF['격리중'] = koreaDF['격리중'].apply(remove_comma)
koreaDF['격리해제'] = koreaDF['격리해제'].apply(remove_comma)
```

```python
koreaDF['확진환자'] = koreaDF['확진환자'].astype(int)
koreaDF['격리중'] = koreaDF['격리중'].astype(int)
koreaDF['격리해제'] = koreaDF['격리해제'].astype(int)
koreaDF['사망자'] = koreaDF['사망자'].astype(int)
koreaDF['발생률'] = koreaDF['발생률'].astype(float)
koreaDF
```

## 데이터 저장

```python
koreaDF.to_csv('korea_corona2.csv')
```

## 확진환자 중 격리중 환자 비율

```python
koreaDF['격리 비율'] = ((koreaDF['격리중'] / koreaDF['확진환자']) *100).round(2)
koreaDF
```

## 확진환자 중 격리해제 환자 비율

```python
koreaDF['격리해제 비율'] = ((koreaDF['격리해제'] / koreaDF['확진환자'])*100).round(2)
koreaDF
```

## 서울이 격리비율이 낮은건 현재 진행중 상태인것으로 예측
## 대구는 격리해제 비율이 높아 처음 코로나 사태이후 안정적인 상태임을 알 수 있음

## 확진환자 중 사망자 환자 비율

```python
koreaDF['사망자 비율'] = ((koreaDF['사망자'] / koreaDF['확진환자']) *100).round(2)
koreaDF
```

## 사망률이 높은 도시

```python
koreaDF.nlargest(5, columns='사망자 비율')
```

## 부산은 발생율에 비해 사망률은 세번째로 높으니 조심해야할 필요가 있음

## 지역별 격리 비율, 격리해제 비율, 사망자 비율 - 라인차트

```python
plt.plot(koreaDF['격리 비율'], 'bo-', 
         koreaDF['격리해제 비율'], 'rs--', 
         koreaDF['사망자 비율'], 'yp-.', 
         alpha=0.5)
plt.xlabel('시도명')
plt.ylabel('비율')
plt.grid(True)
plt.title('지역별 비율')
plt.xticks(rotation=90)
plt.legend(['격리 비율', '격리해제 비율', '사망자 비율'])
plt.show()
```

### 인구대비 확진률

```python
Cp = []
for n1, n2 in zip(ame['인구수'],ame['환자발생수']):
    Cp.append(round(n2/n1,2)*100)
Cp
```
