```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
crimeDF =pd.read_csv('data/crime_in_Seoul.csv',encoding='euc-kr',thousands=',')
crimeDF
```

# 샘플 문제

## 1. 살인발생이 가장높은 관서명, 살인발생,살인검거를 출력하시요

```python
crimeDF['살인 발생'].max()

crimeDF[crimeDF['살인 발생'] == crimeDF['살인 발생'].max()][['관서명','살인 발생','살인 검거']]
```

## 2. 관서별 절도발생,절도검거에 대한 바차트를 그리시요

```python
crimeDF.plot( kind= 'bar', y=['절도 검거','절도 발생'], rot = 0)
plt.show()
```

## 3. 절도검거율 을 구하고 컬럼을 추가 하시요

```python
crimeDF['절도 검거율(%)'] = ((crimeDF['절도 검거']/crimeDF['절도 발생'])*100).round(2)
crimeDF
```

# 3. 관서별 위험도와 위험도 수준 평가
## 기준 - 살인발생, 강도발생, 강간발생의 평균

```python
crimeDF['위험도'] = round((crimeDF['살인 발생']+crimeDF['강도 발생']+crimeDF['강간 발생'])/3, 2)
print(crimeDF['위험도'].quantile(0.33))
print(crimeDF['위험도'].quantile(0.66))
```

```python
crimeDF['위험도 수준'] = crimeDF['위험도'].apply(lambda x:'위험도 매우높음' if x>=71 else '위험도 높음' if x>=43 else '보통')
crimeDF[['관서명','위험도','위험도 수준']].sort_values(by ='위험도', ascending= False)
```

```python
# 상위 5개 위험도 지역
crimeDF[['관서명','위험도','위험도 수준']].sort_values(by ='위험도', ascending= False)[0:5]
```
